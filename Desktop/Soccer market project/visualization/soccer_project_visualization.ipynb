```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

train = pd.read_csv("C:/Users/82103/Desktop/ì¶•êµ¬ ì´ì ì‹œìž¥ í”„ë¡œì íŠ¸/soccer_data.csv")

train.replace([np.inf, -np.inf], np.nan, inplace=True)

def wrangling(train_set):  # ë°ì´í„°ì…‹ ì •ë³´ í™•ì¸ 
    
    print("---Shape---")
    display(train_set.shape)
    
    print("---Info----")
    display(train_set.info())
    
    print("---NaN----")
    display(train_set.isna().sum())
    
    print("---Duplicated---")
    display(train_set[train_set.duplicated()])
    
    print("---Description---")
    display(train_set.describe())
     
    print("---Unique---")
    display(train_set.nunique())
    
wrangling(train)


import math       # í–‰ë ¬ í¬ê¸° ê³„ì‚°ì„ ìœ„í•¨ (ì‹œê°ìžë£Œ í‘œí˜„)

num_cols = train.select_dtypes(exclude='object').columns.tolist()
n = len(num_cols)
cols = 3
rows = math.ceil(n / cols)

plt.figure(figsize=(14, rows * 4))

for i, col in enumerate(num_cols, 1):
    plt.subplot(rows, cols, i)
    sns.histplot(train[col])
    plt.title(col)

plt.tight_layout()


```
